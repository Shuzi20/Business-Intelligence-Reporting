<h2 align="left"><font color= #e67300; font-family= 'Trebuchet MS', sans-serif;>About Dataset:</font></h2>
    
Here is a brief description of each column in the dataset:

1. **Customer_ID**: Unique identifier for each customer.
2. **Customer_Age**: Age of the customer.
3. **Customer_Gender**: Gender of the customer.
4. **Customer_Location**: Geographic location of the customer.
5. **Customer_Type**: Classification of the customer as new or returning.
6. **Order_ID**: Unique identifier for each order.
7. **Order_Date**: Date of the order.
8. **Product_Category**: Category to which the product belongs (e.g., Electronics, Clothing).
9. **Brand**: Brand of the product.
10. **Product_Name**: Name of the product.
11. **Product_ID**: Unique identifier for each product.
12. **Quantity**: Quantity of product ordered.
13. **Product_Price**: Price per unit of the product.
14. **Discount**: Discount applied to the product.
15. **Revenue**: Total revenue from the order after discounts.
16. **Shipping_Fee**: Fee for shipping the order.
17. **Total_Order_Value**: Final order value including shipping and discounts.
18. **Order_Status**: Status of the order (e.g., Completed, Canceled).
19. **Payment_Type**: Payment method used (e.g., Credit Card).
20. **Session_ID**: Unique identifier for each website session.
21. **Page_Views**: Number of pages viewed in the session.
22. **Session_Duration**: Duration of the session in seconds.
23. **Bounce_Rate**: Indicator if session ended without interaction.
24. **Exit_Rate**: Rate of exits from the page.
25. **Campaign_ID**: Unique identifier for each marketing campaign.
26. **Channel**: Marketing channel through which the customer arrived.
27. **Campaign_Cost**: Cost of the marketing campaign.
28. **Coupon_Code**: Code used by customers for discounts.
29. **Ad_Clicks**: Number of clicks on ads in the campaign.
30. **Conversion_Rate**: Percentage of sessions that led to purchases.
31. **Shipping_Method**: Mode of shipping (e.g., Standard, Expedited).
32. **Shipping_Date**: Date the product was shipped.
33. **Delivery_Date**: Date the product was delivered.
34. **Delivery_Status**: Status of the delivery (e.g., On-Time, Delayed).
35. **Repeat_Purchase_Flag**: Indicates if the customer is a repeat buyer.
36. **Customer_Lifetime_Value**: Total revenue from a customer over time.
37. **Fraud_Flag**: Indicator if the transaction was potentially fraudulent.
38. **Risk_Score**: Risk assessment score for the transaction. 

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

#Set display options to show all rows and columns
pd.set_option("display.max_rows", None)  # Show all rows
pd.set_option("display.max_columns",None) # Show all columns
pd.set_option("display.expand_frame_repr",False) # Prevent line breaks in dataframe
```

```python
df = pd.read_csv("ecommerce_bi_dataset.csv")
df.head()
```

```python
# Display the first few rows and the info of the dataset
df.info()
```

```python
df.describe().T
```

<h2 align="left"><font color= #e67300; font-family= 'Trebuchet MS', sans-serif;>Data Cleaning:</font></h2>
    

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display columns with missing values (if any)
print("Missing values per column:")
print(missing_values)
```

```python
# Fill missing values in Coupon_Code with 'No Coupon'
df['Coupon_Code'] = df['Coupon_Code'].fillna('No Coupon')

# Verify if the missing values were handled
missing_values_after_fill = df['Coupon_Code'].isnull().sum()
print(f"\nMissing values in 'Coupon_Code' after filling: {missing_values_after_fill}")
```

```python
# Ensure correct data types for columns
print("\nData types of columns:")
print(df.dtypes)
```

```python
# For datetime columns, ensure they are in datetime64 format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Shipping_Date'] = pd.to_datetime(df['Shipping_Date'])
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])

# Display cleaned data info
print("\nData after cleaning:")
print(df.info())
```

<h2 align="left"><font color= #e67300; font-family= 'Trebuchet MS', sans-serif;>Data Transformation</font></h2>
    

```python
# Extracting the Order Month, Order Year, and Order Day of Week from Order_Date
df['Order_Month'] = df['Order_Date'].dt.month
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Day_of_Week'] = df['Order_Date'].dt.dayofweek  # Monday=0, Sunday=6

# Calculating Profit (Revenue - Shipping Fee - Discount)
df['Profit'] = df['Revenue'] - df['Shipping_Fee'] - df['Discount']

# Verify new columns
print("\nFirst few rows after transformation:")
print(df[['Order_Date', 'Order_Month', 'Order_Year', 'Order_Day_of_Week', 'Profit']].head())
```

<h2 align="left"><font color= #e67300; font-family= 'Trebuchet MS', sans-serif;>Exploratory Data Analysis</font></h2>
    

```python
# Total Revenue and Total Profit
total_revenue = df['Revenue'].sum()
total_profit = df['Profit'].sum()

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Revenue and Profit by Product Category
revenue_by_category = df.groupby('Product_Category')[['Revenue', 'Profit']].sum().reset_index()

# Revenue and Profit by Customer Type
revenue_by_customer_type = df.groupby('Customer_Type')[['Revenue', 'Profit']].sum().reset_index()

# Average Order Value and Average Profit
avg_order_value = df['Total_Order_Value'].mean()
avg_profit = df['Profit'].mean()

print(f"\nAverage Order Value: ${avg_order_value:,.2f}")
print(f"Average Profit: ${avg_profit:,.2f}")
```

```python
# Display the summarized data for further insights
print("\nRevenue and Profit by Product Category:")
revenue_by_category
```

```python
print("\nRevenue and Profit by Customer Type:")
revenue_by_customer_type
```

<h2 align="left"><font color= #e67300; font-family= 'Trebuchet MS', sans-serif;>Identifying Key Matrics</font></h2>
    

```python
# Display the calculated KPIs
# Customer Lifetime Value (CLV) - Total Revenue / Number of Unique Customers
unique_customers = df['Customer_ID'].nunique()
clv = total_revenue / unique_customers
print(f"\nCustomer Lifetime Value (CLV): ${clv:,.2f}")
```

```python
# Conversion Rate (Total Orders / Total Page Views)
total_page_views = df['Page_Views'].sum()
total_orders = df['Order_ID'].nunique()
conversion_rate = total_orders / total_page_views
print(f"Conversion Rate: {conversion_rate:.2%}")
```

```python
# Customer Retention Rate (percentage of customers with Repeat_Purchase_Flag = 1)
repeat_customers = df[df['Repeat_Purchase_Flag'] == 1]['Customer_ID'].nunique()
retention_rate = (repeat_customers / unique_customers) * 100
print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

```python
# Export the cleaned and transformed dataset to a CSV file
df.to_csv('ecommerce_data_transformed.csv', index=False)

print("\nData has been successfully exported to 'ecommerce_data_transformed.csv'")
```
